```python
import random
x=1.0
y=1.0

delta=0.1

N_hits=0
N_trials=4000

for i in range(N_trials):
    del_x,del_y=random.uniform(-delta,delta),random.uniform(-delta,delta)
    if abs(x+del_x)<1 and abs(y+del_y)<1:
        x+=del_x
        y+=del_y
    if x**2 +y**2 <1:
        N_hits+=1

print("value of pi",4*N_hits/float(N_trials))
```

```python

def markov_pi(N):
    x=1.0
    y=1.0

    delta=0.1

    N_hits=0
    
    for i in range(N):
        del_x,del_y=random.uniform(-delta,delta),random.uniform(-delta,delta)
        if abs(x+del_x)<1 and abs(y+del_y)<1:
            x+=del_x
            y+=del_y
        if x**2 +y**2 <1:
            N_hits+=1
    return N_hits

N_trials=4000
N_times=100

for i in range(N_times):
    print("value of pi = ", 4*markov_pi(N_trials)/float(N_trials))   
```
