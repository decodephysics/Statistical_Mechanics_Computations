```python
import random

N_hits=0
N_trials=4000

for i in range(N_trials):
    x,y=random.uniform(-1,1),random.uniform(-1,1)
    if x**2 + y**2 < 1:
        N_hits+=1

print("value of pi is",4*N_hits/float(N_trials))
```

```python
def direct_pi(N):
    N_hits=0
    for i in range(N):
        x,y=random.uniform(-1,1),random.uniform(-1,1)
        if x**2 + y**2 < 1:
            N_hits+=1
    return N_hits 

N_times=100
N_trials=4000

for i in range(N_times):
    print("value of pi",4*direct_pi(N_trials)/float(N_trials))
```
